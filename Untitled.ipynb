```python
import numpy as np
```

```python
X = np.array([[3, -2],[-1, 4]])
```

```python
XT = X.T
```

```python
XT
```

```python
XTX = XT.dot(X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([[-4],[1]])
hat = XTX_inv.dot(XT).dot(y)
hat
```

```python
import matplotlib.pyplot as plt
# x = np.linspace(-20, 20, 1000)
x = np.arange(100)
X = np.sin(x)
Y = x * np.cos(x)
Z = x

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(X, Y, Z)

ax.set_xlabel('sin(x)')
ax.set_ylabel('x cos(x)')
ax.set_zlabel('x')

plt.show()
```
