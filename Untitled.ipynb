```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grid
xx, yy = np.meshgrid(
    np.linspace(-10, 10, 30),
    np.linspace(-10, 10, 30)
)

# Plane equation: c0*x + c1*y + c2*z = 0    z = (-c0*x - c1*y)/c2
z1 = (-c[0] * xx - c[1] * yy) / c[2]

# Plot surface
ax.plot_surface(xx, yy, z1, alpha=0.7)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.plot([0, a[0]], [0, a[1]], [0, a[2]], 'k')
ax.plot([0, b[0]], [0, b[1]], [0, b[2]], 'k')
ax.plot([0, c[0]], [0, c[1]], [0, c[2]], 'r')
# ax.view_init(azim=150, elev=45)
plt.show()
      
```

```python
import numpy as np
import matplotlib.pyplot as plt

s1 = np.array([1, 1, 0])
s2 = np.array([1, 7, 0])

v = np.array([1, 2, 0])
w = np.array([3, 2, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# vectors
ax.plot([0, s1[0]], [0, s1[1]], [0, s1[2]], 'r')
ax.plot([0, s2[0]], [0, s2[1]], [0, s2[2]], 'r')
ax.plot([0, v[0]],  [0, v[1]],  [0, v[2]],  'k')
ax.plot([0, w[0]],  [0, w[1]],  [0, w[2]],  'k')

# plane from s1 and s2
cp = np.cross(s1, s2)

xx, yy = np.meshgrid(np.linspace(-15,15,30),
                     np.linspace(-15,15,30))

z1 = (-cp[0]*xx - cp[1]*yy) / cp[2]
ax.plot_surface(xx, yy, z1, alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
